# Histograma
### Son usualmente el primer gráfico a realizar de una columna numérica.

## Bins: Rangos equidistantes de agrupaciones de valores numéricas.
### Consejos: 
### Todos los bins tienen que tener exactamente el mismo tamaño.
### No tiene sentido hacer un histograma de una variable categórica. Esto sería un diagrama de barras común y corriente.

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
df.hist(
    'Age',
    by='Sex',
    bins=15
)
```
