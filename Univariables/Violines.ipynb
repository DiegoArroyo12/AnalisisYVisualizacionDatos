# Gráfico de Violines
### Manteniendo todas las ventajas de un diagrama de cajas, me muestra la forma de la distribución de mi variable.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
sns.violinplot(
    df,
    x='Sex',
    y='Age'
    #inner='quartiles' Mostramos los cuartiles
)
# Tiene un diagrama de caja, en donde el punto blanco simboliza la mediana
```
