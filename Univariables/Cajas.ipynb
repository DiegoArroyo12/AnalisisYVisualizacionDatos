# Gráfico de Cajas
### Muestra estadísticos de una variable de uno o más grupos de mis datos mediante cajas y líneas.

### Se utiliza cuando quiero comparar diversos grupos a través de sus estadísticos (media, mediana y cuartiles).

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
# Le decimos que busque en la columna de Age y elimine los campos vacíos por filas (axis)
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
df.boxplot(
    column='Age',
    by='Sex'
)
```

```python
import seaborn as sns
```

```python
sns.boxplot(
    df,
    x='Sex',
    y='Age'
)
```
