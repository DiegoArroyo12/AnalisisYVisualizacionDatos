# Gráfico de Densidad (KDE)
## Kernel Density Estimator
### Es un "suavizado" del histograma que tiene la forma de la distribución de los valores numéricos. Formalmente es una función de probabilidad.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
sns.histplot(
    df,
    x='Age',
    kde=True,
    bins=20,
    hue='Sex' # Crea dos histogramas
    #stat='probability' # Función de probabilidad
)
```

```python
sns.histplot(
    df,
    x='Age',
    cumulative=True,
    bins=20,
    fill=False,
    element='step' # Deja solo los escalones
)
```
